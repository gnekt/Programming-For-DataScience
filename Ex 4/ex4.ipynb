Define a class "City" with attributes the following information about a city:
- attribute: Name of the city
- attribute: Number of inhabitants

Define a class "Region" with  
- attibute: Name of the region
- attribute: An arbitrary number of cities
- method: to get the number of cities
- method: to get the total number of inhabitants of the region

Define a class "Nation" with
- attribute: Name of the nation
- attribute: An arbitrary number of regions
- method: to get the number of regions
- method: to get the number of cities
- method: to get the total number of inhabitants of the nation

```python
from typing import List
class City():
    """
    Define a class "City" with attributes the following information about a city:

        attribute: Name of the city
        attribute: Number of inhabitants
    """
    def __init__(self,city_name: str, number_inhabitants: int):
        self.city_name = city_name 
        self.number_inhabitants = number_inhabitants
    
    def __str__(self):
        return f"The city is {self.city_name} and has {self.number_inhabitants} inhabitants"
        
class Region():
    """
    Define a class "Region" with

        attibute: Name of the region
        attribute: An arbitrary number of cities
        method: to get the number of cities
        method: to get the total number of inhabitants of the region
    """
    def __init__(self, region_name: str, cities: List[City]):
        self.region_name = region_name
        self.cities = cities
        
    def get_number_cities(self):
        """
            Get the number of cities
        """
        return len(self.cities)
    
    def get_total_numer_inhabitants(self):
        """
            Get the total number of inhabitants of the region
        """
        return sum(city.number_inhabitants for city in self.cities) 
    
    def __str__(self):
        return f"The Region is {self.region_name}, it has {self.get_number_cities()} cities and {self.get_total_numer_inhabitants()} inhabitants in total"

class Nation():
    """
    Define a class "Nation" with

        attribute: Name of the nation
        attribute: An arbitrary number of regions
        method: to get the number of regions
        method: to get the number of cities
        method: to get the total number of inhabitants of the nation
    """
    def __init__(self, nation_name: str, regions: List[Region]):
        self.nation_name = nation_name
        self.regions = regions
    
    def get_number_regions(self):
        """
            Get the number of regions
        """
        return len(self.regions)
    
    def get_number_cities(self):
        """
            Get the number of cities
        """
        return sum(region.get_number_cities() for region in self.regions)
    
    def get_total_number_inhabitants(self):
        """
            Get the total number of inhabitants of the nation
        """
        return sum(region.get_total_numer_inhabitants() for region in self.regions)
    
    def __str__(self):
        total_str = f"The Nation is {self.nation_name}, the regions are: \n"
        for region in self.regions:
            total_str += f"{region}\n"
        total_str += f"with a {self.get_total_number_inhabitants()} total number of inhabitants"
        return total_str
```

## Step 4

Overload some operators.

For the class City:

 - "<": to check if a city has less inhabitants then another
 - "<=": to check if a city has less inhabitants then another
 
 
 For the class Region:
 
 - "in": to check if a city belongs to a given region
 
 
 For the class Nation
 
  - "in": to check if a region OR city belongs to the given region

```python
from typing import List
class City():
    """
    Define a class "City" with attributes the following information about a city:

        attribute: Name of the city
        attribute: Number of inhabitants
    """
    def __init__(self,city_name: str, number_inhabitants: int):
        self.city_name = city_name 
        self.number_inhabitants = number_inhabitants
    
    def __str__(self):
        return f"The city is {self.city_name} and has {self.number_inhabitants} inhabitants"
    
    def __lt__(self,oth):
        return self.number_inhabitants < oth.number_inhabitants
    
    def __leq__(self,oth):
        return self.number_inhabitants <= oth.number_inhabitants
         
class Region():
    """
    Define a class "Region" with

        attibute: Name of the region
        attribute: An arbitrary number of cities
        method: to get the number of cities
        method: to get the total number of inhabitants of the region
    """
    def __init__(self, region_name: str, cities: List[City]):
        self.region_name = region_name
        self.cities = cities
        
    def get_number_cities(self):
        """
            Get the number of cities
        """
        return len(self.cities)
    
    def get_total_numer_inhabitants(self):
        """
            Get the total number of inhabitants of the region
        """
        return sum(city.number_inhabitants for city in self.cities) 
    
    def __str__(self):
        return f"The Region is {self.region_name}, it has {self.get_number_cities()} cities and {self.get_total_numer_inhabitants()} inhabitants in total"
    
    def __contains__(self,oth: City):
        return oth in self.cities

class Nation():
    """
    Define a class "Nation" with

        attribute: Name of the nation
        attribute: An arbitrary number of regions
        method: to get the number of regions
        method: to get the number of cities
        method: to get the total number of inhabitants of the nation
    """
    def __init__(self, nation_name: str, regions: List[Region]):
        self.nation_name = nation_name
        self.regions = regions
    
    def get_number_regions(self):
        """
            Get the number of regions
        """
        return len(self.regions)
    
    def get_number_cities(self):
        """
            Get the number of cities
        """
        return sum(region.get_number_cities() for region in self.regions)
    
    def get_total_number_inhabitants(self):
        """
            Get the total number of inhabitants of the nation
        """
        return sum(region.get_total_numer_inhabitants() for region in self.regions)
    
    def __str__(self):
        total_str = f"The Nation is {self.nation_name}, the regions are: \n"
        for region in self.regions:
            total_str += f"{region}\n"
        total_str += f"with a {self.get_total_number_inhabitants()} total number of inhabitants"
        return total_str
    
    def __contains__(self,oth):
        if isinstance(oth, City): return any(oth in region for region in self.regions)
        return oth in self.regions
        
```
