* function2 uses a cycle along the columns and np.mean for calculating the averages
* function3 uses np.mean with the axis parameter (and no cycle)

Use the %timeit directive to estimate the perfomances of the three functions

```python
X = np.random.uniform(low = 0, high = 1, size = (1000,1000))
```

```python
def mean_columns_method_one(X):
    mean = []
    for column in X.T:
        mean.append(sum(column)/len(column))
    return mean

def mean_columns_method_two(X):
    mean = []
    for column in X.T:
        mean.append(np.mean(column))
    return mean

def mean_columns_method_three(X):
    M = np.mean(X,axis=1)
```

# Problem 4

Define a function that takes as input an (n x 3) array, with n being the number of samples and 3 the number of dimensions, and that it returns an (n x n) array  with element i,j equal to the distance between sample i and sample j. First, do it using any cycle you want. Then, try to get rid of the cycles, until none is left. Compare the performaces of the different implementations, using random data sets of different size.

```python
n = 100
A = np.random.uniform(low = 0, high = 1, size = (n,3))
```

```python
def method_1(A,n):
    D = np.empty((A.shape[0],A.shape[0]))
    for outer_iterator in range(A.shape[0]):
        for inner_iterator in range(A.shape[0]):
            D[outer_iterator,inner_iterator] = np.sum( (A[outer_iterator,:] - A[inner_iterator,:])**2) **.5
            
def method_2(A):
    D = np.empty((A.shape[0],A.shape[0]))
    for iterator in range(A.shape[0]):
        D[iterator,:] = np.sum( (A - A[iterator,:].reshape(1,-1)) **2) **.5      

def method_3(A):
    return np.sum(
        (A.reshape((A.shape[0],A.shape[1],1))
         -A.transpose().reshape((1,A.shape[1],A.shape[0])))**2, axis = 1)**0.5
            
%timeit method_1(A,n)
%timeit method_2(A)
%timeit method_3(A)
```
