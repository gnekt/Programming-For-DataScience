* Define a class to iterate over numbers of the series n(n+1)/2 up to a maximum value max_n

```python
class my_series():
    def __init__(self, max_n=0):
        self.max_n = max_n

    def __iter__(self):
        self.n = 0
        return self

    def __next__(self):
        if self.n <= self.max_n:
            result = (self.n*(self.n+1))/2
            self.n += 1
            return result
        else:
            raise StopIteration

series = my_series(20)
for x in series:
    print(x)
```

* Define a class to iterate over prime numbers up to a maximum value max_n. Calculate the prime numbers at run-time by finding the next number that can not be divided by any lower number higher than 1 (Hint: remember break/else in cycle)

```python
class my_prime_number_series():
    def __init__(self, max_n=0):
        self.max_n = max_n

    def __iter__(self):
        self.n = 1
        return self

    def __next__(self): 
        while True:
            prime_flag = True
            if self.n % 2 == 0 or self.n == 1: 
                self.n += 1
                continue
            
            for i in range(2,self.n): 
                if self.n % i == 0: 
                    prime_flag = False
                    self.n += 1
                    break
            if prime_flag: 
                oth = self.n 
                self.n += 1
                break
        if oth > self.max_n: raise StopIteration
        return oth
        

series = my_prime_number_series(20)
for x in series:
    print(x)
```

* Conver the previous two classes to generators

```python
def triangular(max_n=100):
    n=0
    while n < max_n:
        n +=1
        yield (n*(n+1))/2

for i in triangular(10):
      print(i)
```

```python
def prime(man_n):
    n = 2
    while n < man_n:
        prime_flag = True
        if n % 2 == 0 : 
            n += 1
            continue
        
        for i in range(2,n): 
            if n % i == 0: 
                prime_flag = False
                n += 1
                break
        if prime_flag: 
            yield n
        n += 1
    
    
   

for p in prime(20):
    print(p) 
```
