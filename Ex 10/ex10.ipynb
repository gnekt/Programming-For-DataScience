```python
import pandas as pd
import numpy as np
```

* The file ex10_1.txt includes the names of the 20-th NBA players that scored more points
* Define a Series with index the name of the players and values the total number of points


```python
with open("ex10_1.txt","r") as nba:
    lines = nba.readlines()
    scores = pd.Series([int(point.split("\t")[2].replace("\n","")) for point in lines[1:]], index=[point.split("\t")[1] for point in lines[1:]])
print(scores)
```

* The file ex10_2.txt includes the names of the 20-th NBA players that played more games
* Define a Series with index the name of the players and values the total number of games played

```python
with open("ex10_2.txt","r") as nba:
    lines = nba.readlines()
    games = pd.Series([int(game.split()[4].replace("\n","")) for game in lines[1:]], index=[point.split("\t")[1] for point in lines[1:]])
print(games)
```

* Print the number of points scored by Tim Duncan
* Print the number of points scored by the 3 top scorers

```python
print(scores['Tim Duncan'])
print(scores[:3])
```

* Define a new series that includes only the players that scored more than 30000 points

```python
score_more_than = scores[scores > 30000]
print(score_more_than)
```

* Remove Dominique Wilkins from the top scoreres

```python
series_dropped = scores.drop('Dominique Wilkins')
print(series_dropped)
```

* Define a new Series equal to the average point per game
* Use the method dropna to remove players that are not in one of the two Series
* Use the method sort_values to sort the series

```python
averages = scores / games
print(averages.dropna().sort_values())
```

* Merge the series of points and games into a pd.DataFrame

```python
nba = pd.DataFrame({'games':games,"scores":scores})
nba.index.name = 'Name'
print(nba)
```

* Add to the dataframe a column with the average points per game using the pd.Series defined above

```python
nba['average_scores_games'] = averages
print(nba)
```

* Add to the dataframe the data of the following players:
    * Charles Barkley, points = 23757, games = 1073
    * Pau Gasol, points = 20894, games = 1226

```python
nba = nba.append(pd.DataFrame({
    'scores':{'Charles Barkley':23757, 'Pau Gasol':20894},
    'games':{'Charles Barkley':1073, 'Pau Gasol':1226},
}))
print(nba)
nba['average_scores_games'] = nba.scores/nba.games
```

* Select only the rows of the dataframe corresponding to players with average number of points per game above 25.0
* Select only the rows of the dataframe corresponding to players with average number of points per game above 20.0 and number of games above 1300

```python
print(nba[nba['average_scores_games'] > 25])
print(nba[(nba.average_scores_games > 20.0) & (nba.games > 1300)])
```
