# Problem 1

The file *ex3_data_clean.csv* in the current directory contains the results of a sequencing experiment for chromosome 1 of three subjects, with the following format:

*2,"248487768","248487768","OR2M7","23.3","0.800919"*

where:
* 2 is the id of the subject
* "248487768" is the starting position of the variant
* "248487768" is the ending position of the variant
* "OR2M7" is the name of the mutated gene
* "23.3" is the cadd estimate of pathogenicity (a bioinformatic algorithm)
* "0.800919" is the frequency of the variant in the general population

Write the code for reading the file, and storing the data into 6 lists, containing respectivelly the subject id (as a str), the starting position (as an int), the ending positions (as an int), the gene name (as a str), the cadd value (as a float), the frequency (as a float).

```python
id_list = []
starting_position_list =[]
ending_position_list = []
gene_name_list = []
cadd_value_list = []
frequency_list = []
with open("ex3_data_clean.csv","r") as input_file:
    for line in input_file:
        gene_id,starting_position,ending_position,gene_name,cadd_value,frequency = line.split(",")
        id_list.append(str(gene_id))
        starting_position_list.append(int(float(starting_position.replace("\"","").replace("'",""))))
        ending_position_list.append(int(float(ending_position.replace("\"","").replace("'",""))))
        gene_name_list.append(str(gene_name))
        cadd_value_list.append(float(cadd_value.replace("\"","").replace("'","")))
        frequency_list.append(float(frequency.replace("\"","").replace("'","")))
print(frequency_list)
```

* Create a new dictionary (rare_variants) with key the subject ids and with values a list of tuples, including the starting/ending position of all the variants with frequency < 0.1
* How many rare variants (frequency < 0.1) are present for each subject ?

The file *ex3_data_clean.csv* in the current directory contains the results of a sequencing experiment for chromosome 1 of three subjects, with the following format:

*2,"248487768","248487768","OR2M7","23.3","0.800919"*

where:
* 2 is the id of the subject
* "248487768" is the starting position of the variant
* "248487768" is the ending position of the variant
* "OR2M7" is the name of the mutated gene
* "23.3" is the cadd estimate of pathogenicity (a bioinformatic algorithm)
* "0.800919" is the frequency of the variant in the general population

```python
rare_variant = {}
with open("ex3_data_clean.csv","r") as input_file:
    for line in input_file:
        gene_id,starting_position,ending_position,gene_name,cadd_value,frequency = line.split(",")
        if float(frequency.replace("\"","").replace("'","")) > 0.1: continue
        try:
            rare_variant[gene_id].append((starting_position,ending_position))
        except KeyError as ex:
            rare_variant[gene_id] = []
            rare_variant[gene_id].append((starting_position,ending_position))
for index,data in rare_variant.items():
    print(f"Subject: {index}, total number: {len(data)}")
```

* Identify the number of rare variants in common between all the possible pairs of subjects

```python
for subj_i, rare_i in rare_variant.items():
    for subj_j, rare_j in rare_variant.items():
        print(subj_i, subj_j, len(set(rare_i) & set(rare_j)))   
```

# Problem 2 

Write a function that creates the same dictionary as before, but reading from the file *ex3_data.csv*. The difference with *ex3_data_clean.csv* is that in *ex3_data.csv* there are many lines with missing fields or with fields that contain values of the wrong type (e.g. a frequency that is not a number). The dictionary should include only the lines that have valid values for all the fields.

```python
gene_list = {
    "id_list" : [],
    "starting_position_list" :[],
    "ending_position_list" : [],
    "gene_name_list" : [],
    "cadd_value_list" : [],
    "frequency_list" : []
}

with open("ex3_data.csv","r") as input_file:
    for line in input_file:
        data = line.split(",")
        if len(data) != 6:continue
        if any(dato == '' or dato == "" for dato in data): 
            continue
        data = [dato.strip().replace("\"","").replace("'","") for dato in data]
        if not(data[0].isdigit()) or not(data[1].isdigit()) or not(data[2].isdigit()) or not(data[4].isdigit()) or not(data[5].replace(".","").isdigit()):continue
        try:
            gene_list["id_list"].append(str(data[0]))
            gene_list["starting_position_list"].append(int(data[1]))
            gene_list["ending_position_list"].append(int(data[2]))
            gene_list["gene_name_list"].append(str(data[3]))
            gene_list["cadd_value_list"].append(float(data[4].replace("\"","").replace("'","")))
            gene_list["frequency_list"].append(float(data[5].replace("\"","").replace("'","")))
        except Exception as e:
            print(e)
print(gene_list)
```
