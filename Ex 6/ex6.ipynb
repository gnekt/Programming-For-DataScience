```python
class Vehicle(object):
    def __init__(self, name, km = 0, num_wheels = None):
        self.name = name
        self.km = km
        self.num_wheels = num_wheels
    
    def add_km(self,*args):
        if any(type(arg) is not float for arg in args) or any(type(arg) is not int for arg in args):
            raise TypeError
        summation_km = sum(arg for arg in args)
        self.km += summation_km
    
    def __str__(self):
        return f"Car: {self.name}, kilometers travelled: {self.km}"
```

# Problem 2

Given the class "Point", which represents a point in an n-dimensional space

```python
class Point(object):
    
    def __init__(self, *args, label=''):
        self.coords = [x for x in args]
        self.label = label
        
    def __str__(self):
        return " ".join(str(x) for x in self.coords)
    
    def __len__(self):
        return len(self.coords)
```

## Step 1

Define the class "Vector" extending the class "Point".


For the class "Vector"
- define a method to compute the module of the vector (euclidean distance from the origin). Hint: from math library import sqrt to use the square root function.
- overload the operator " < " to know if a vector has a smaller module then other. 
- overload the operator " + " to compute the vector sum of two vectors.

Note: it make sense to compare and sum only vectors having same dimensional space.

Define a few vectors to test the class

```python
from math import sqrt
class Vector(Point):

    def module(self):
        """
            module: Compute the euclidean distance from the origin
        """
        return sqrt(sum([coordinate**2 for coordinate in self.coords]))
    
    def __lt__(self, other):
        if not isinstance(other,Vector): raise TypeError
        if len(self) != len(other): raise ValueError("Different dimension of other vector")
        return self.module() < other.module()
    
    def __add__(self, other): 
        if not isinstance(other,Vector): raise TypeError
        if len(self) != len(other): raise ValueError("Different dimension of other vector")
        return Vector([coordinate_1 + coordinate_2 for (coordinate_1,coordinate_2) in zip(self.coords, other.coords)],label=f"{self.label}+{other.label}")

# Define two vectors (same dimension)
A = Vector(2, 0, 0, label='A')
B = Vector(1, 0, 0, label='B')

# Compare their modules
print("A has a smaller module then B:", A < B)

# and compute the sum
C = A + B
print("A + B =", C)
```

## Step 1

Define the class "Bot". It must include:
- a mandatory attribute for the name;
- an attribute for the position in the room. (two coordinates, default: 0 0)
- define __ str __ () in order to print the bot name and position.

```python
class Bot:
    
    def __init__(self, name, coordinates = [0,0]):
      self.name = name
      self.coordinates = coordinates
    
    def __str__(self):
        return f"The Bot {self.name} is in position: {self.coordinates}"
```

## Step 2

Extend the class "Bot" in order to define "MovingBot". A moving bot is able to modify its position. 
- define a method to make a movement. It receives two arguments which indicate the correspondent movent in the x and in the y axes. If no arguments are passed, the moving bot takes a random movement (in the range of -5, 5 for each coordinate). Hint: from random library import randint(a, b) to get a random integer from a to b.

```python
from random import randint
class MovingBot(Bot):
    
    def movement(self, x = randint(-5,5), y = randint(-5,5)):
        old_coordinates = self.coordinates
        self.coordinates = [self.coordinates[0]+x,self.coordinates[1]+y]
        print(f"Robot move: from {old_coordinates} to {self.coordinates}")
```
